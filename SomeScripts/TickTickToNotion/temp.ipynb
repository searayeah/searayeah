```python
import pandas as pd
```

```python
df = pd.read_csv("TickTick-backup-2022-05-11.csv")
```

```python
df.columns
```

```python
df["Status"] = df["Status"].apply(lambda x: True if x in [1, 2] else False)
```

```python
df_Repeat = df[df["List Name"] == "Repeat"].copy()
df_Repeat.drop(["Folder Name", "List Name", "Column Name"] +     [
        "Tags",
        "Content",
        "Is Check list",
        "Start Date",
        "Due Date",
        "Reminder",
        "Repeat",
        "Priority",
        "Created Time",
        "Order",
        "Timezone",
        "Is All Day",
        "Is Floating",
        "Column Order",
        "View Mode",
        "taskId",
        "parentId",
    ], axis=1).to_csv("Repeat.csv",index=False)
df_Repeat.head()
```

```python
df.drop(
    [
        "Tags",
        "Content",
        "Is Check list",
        "Start Date",
        "Due Date",
        "Reminder",
        "Repeat",
        "Priority",
        "Created Time",
        "Completed Time",
        "Order",
        "Timezone",
        "Is All Day",
        "Is Floating",
        "Column Order",
        "View Mode",
        "taskId",
        "parentId",
    ],
    inplace=True,
    axis=1,
)
df.head()
```

```python
# DS roadmap
ds_roadmap = df[df["Folder Name"] == "DS ROADMAP"].copy()
ds_roadmap.drop(["Folder Name"], axis=1).to_csv("DS ROADMAP.csv", index=False)
ds_roadmap.head()
```
